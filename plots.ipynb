```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot(game_data, game_number, title=None):
    game_data = dict(sorted(game_data.items(), key=lambda x: x[1], reverse=True))

    sns.set(style="whitegrid")
    plt.figure(figsize=(10, 6))
    sns.barplot(x=list(game_data.values()), y=list(game_data.keys()), palette="viridis")
    for i, v in enumerate(game_data.values()):
        plt.text(v + 3, i + 0.25, str(v), color="black", fontweight="bold")
    if title:
        plt.title(title)
    else:
        plt.title(f"Game {game_number} results")
    plt.xlabel("Scores")
    plt.ylabel("Players")
    if title:
        plt.savefig(f"{title}.png", dpi=500)
    else:
        plt.savefig(f"game_{game_number}.png", dpi=500)
    plt.show()
```

```python
game1 = {
    "Henrique": 12,
    "Rodrigo": 180,
    "Oscar": 84,
    "Edilson": 63,
    "Rodolfo": 162,
    "Hagi": 105,
    "Victor": 69,
    "Andre": 165,
}

plot(game1, 1)
```
